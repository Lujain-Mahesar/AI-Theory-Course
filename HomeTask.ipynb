```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

#Can we remove outliers or apply normalization/standardization on complete table?
We cannot remove outliers or apply normalization/standardization on complete table because it contains both numerical columns and categorical columns. Columns like Age and Fare are numerical and can be scaled, but columns like Sex, Embarked, or Survived are categorical. If we wanted to include categorical columns, we would first need to convert them into numerical form (for example, using encoding). Otherwise, applying outlier removal or normalization to the entire dataset will mess up with the data and give incorrect results.So the correct approach is to handle numerical columns separately for scaling or outlier removal, and handle categorical columns appropriately (encoding or leaving them as they are).

#What is the difference between the Pandas series and dataframe?
A pandas series is a one-dimensional labeled array that can hold any type of data, such as integers, floats, or strings. Each value in a Series has an index that identifies it. You can think of a Series as a single column of a table.

A pandas dataframe is two-dimensional, like a table or spreadsheet. It has rows and columns, where each column is a Series. DataFrames can store multiple types of data across columns, such as numbers, strings, or dates.

A series represents a single column, while a dataframe is a table consisting of multiple columns.

#Axis Explanation

Axis tells Pandas whether we are working with rows or columns.  
axis=0 means rows. For example, if we want to drop a row, we use axis=0.  
axis=1 means columns. For example, if we want to drop a column or do calculations on a column, we use axis=1.  

Basically, it’s just a way to tell Pandas the direction to apply an operation.

#What is the sort function in pandas?
In Pandas, the sort_values() function is used to sort the data in a column. We can sort in ascending or descending order depending on what we need. Sorting helps us organize the data, for example to see the highest or lowest values in a column.

```python
#Sorting the fare column in ascending order
df.sort_values('Fare', ascending=True)
```

```python
# Sorting Fare in descending order
df.sort_values('Fare', ascending=False)
```

#How do I access data in a dataframe?
In pandas, we can access data using loc or iloc.
- loc is label-based. We use the row and column labels
- iloc is position-based. We use row and column numbers.


```python
df.loc[0, 'Age']  # value in row 0, column 'Age'
```

```python
df.loc[0:5, ['Name','Fare']]  # rows 0 to 5, columns Name and Fare
```

```python
df.iloc[0,3] # value in row 0, column at position 3
```

```python
df.iloc[0:5, 0:3] # rows 0 to 4 (5 not included), columns 0 to 2 (3 not included)
```
